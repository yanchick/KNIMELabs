```python
import tflearn
```

```python
import numpy
```

```python
import matplotlib.pyplot as plt
```

```python
x0 = numpy.transpose([numpy.arange(0,6.28,0.01)])
x= x0[:,0].tolist()
print x
```

```python
y = numpy.transpose([numpy.sin(x0).tolist()])[0]
y = y[:,0].tolist()
print y
```

```python
plt.plot(x,y)
plt.show()
```

```python
net = tflearn.input_data(shape=[None,1])
net = tflearn.fully_connected(net,10,activation='tanh')
net = tflearn.fully_connected(net,1,activation='linear')
```

```python
regression = tflearn.regression(net, optimizer='sgd', loss='mean_square',
                                 metric='R2', learning_rate=0.01)
```

```python
m = tflearn.DNN(regression)
```

```python
m.fit(x, y, n_epoch=10, show_metric=True, snapshot_epoch=False)
```

```python
print y
```

```python
m.predict([[1]])
```
